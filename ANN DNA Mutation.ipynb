```python
import pandas as pd

data = pd.read_csv('Documents/Data Mutasi Covid-19/data_mutasi_baru_variasi_1.csv', ';')
data = pd.read_csv('Documents/Data Mutasi Covid-19/Data_mutasi_separasi_1.csv', ';')

data
```

```python
data
```

```python
dataset = data.values
dataset
```

```python
X = dataset[:,0:42]
Y = dataset[:,43]

print(X)
print(Y)
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
X_scale = min_max_scaler.fit_transform(X)

X_scale
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X_scale, Y, test_size=0.2)
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.8)

print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential([
    Dense(128, activation='relu', input_shape=(42,)),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(256, activation='relu'),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')])

model.summary()
```

```python
import keras
opt = keras.optimizers.Adam(lr=0.001)

model.compile(optimizer=opt,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,
                 batch_size=32, 
                 epochs=10,
                 validation_data=(X_test, Y_test))
```

```python
hasil  = model.evaluate(X_test, Y_test)[1]
print("Akurasi Training:", hasil*100, "%")
```

## Evaluation

## Machine Learning Method

```python
# Importing Classifier Modules
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

import numpy as np

import pandas as pd
```

```python
#METODE
k_fold = KFold(n_splits=10, shuffle=True, random_state=1)

#DecisionTree
clf1 = DecisionTreeClassifier()
scoring = 'accuracy'
clf1_fit = clf1.fit(X, Y)
score1 = cross_val_score(clf1, X, Y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score1*100)

# decision tree Score
a1 = round(np.mean(score1)*100, 2)
print("Akurasi DT :", a1, "%")
```

```python
#KNN
clf2 = KNeighborsClassifier(n_neighbors = 2)
scoring = 'accuracy'
clf2_fit = clf2.fit(X, Y)
score2 = cross_val_score(clf2, X, Y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score2*100)
# kNN Score
b1 = round(np.mean(score2)*100, 2)
print("Akurasi KNN :", b1, "%")
```

```python
#Random  Forest
clf3 = RandomForestClassifier(n_jobs=2, random_state=0)
scoring = 'accuracy'
clf3_fit = clf3.fit(X, Y)
score3 = cross_val_score(clf3_fit, X, Y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score3*100)

#Random Forest Score
#clf.fit(train[features], y)
c1 = round(np.mean(score3)*100, 2)
print("Akurasi Random Forest :", c1, "%")
```

```python
#SVM
#clf4 = RandomForestClassifier(n_jobs=2, random_state=0)
#Create a svm Classifier
clf4 = svm.SVC(kernel='linear') # Linear Kernel
scoring = 'accuracy'
clf4_fit = clf4.fit(X, Y)
score4 = cross_val_score(clf4_fit, X, Y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score4*100)

#SVM Score
d1 = round(np.mean(score4)*100, 2)
print("Akurasi SVM :", d1, "%")
```

```python
#Gaussian NB
#Create a Gaussian Classifier
clf5 = GaussianNB()
scoring = 'accuracy'
clf5_fit = clf5.fit(X, Y)
score5 = cross_val_score(clf5_fit, X, Y, cv=k_fold, n_jobs=1, scoring=scoring)
print(score5*100)

#NB Score
e1 = round(np.mean(score5)*100, 2)
print("Akurasi SVM :", e1, "%")
```
